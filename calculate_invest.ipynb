```python
# Agrupar por Marca, Semana y Año, y sumar el costo
resumen_por_marca_semana_año = df_meta.groupby(['Marca', 'Semana', 'Año'])['Cost'].sum().reset_index()
resumen_por_marca_semana_año = resumen_por_marca_semana_año.sort_values(['Año', 'Semana', 'Cost'], ascending=[True, True, False])

# Mostrar el resultado
print("Suma de Costo por Marca, Semana y Año:")
print(resumen_por_marca_semana_año)

# También calculamos el resumen solo por marca para comparación
resumen_por_marca = df_meta.groupby('Marca')['Cost'].sum().reset_index()
resumen_por_marca = resumen_por_marca.sort_values('Cost', ascending=False)
print("\nSuma de Costo por Marca (total):")
print(resumen_por_marca)

# También calculamos el resumen por marca y año
resumen_por_marca_año = df_meta.groupby(['Marca', 'Año'])['Cost'].sum().reset_index()
resumen_por_marca_año = resumen_por_marca_año.sort_values(['Año', 'Cost'], ascending=[True, False])
print("\nSuma de Costo por Marca y Año:")
print(resumen_por_marca_año)

# Mostramos el dataframe original para referencia
df_meta
```

```python
# Filtrar para la semana 12
semana_12 = resumen_por_marca_semana_año[resumen_por_marca_semana_año["Semana"] == 12]

# Mostrar el dataframe filtrado
print("Datos de la semana 12:")
print(semana_12)

# Calcular y mostrar el total de Cost para la semana 12
total_cost_semana_12 = semana_12['Cost'].sum()
print(f"\nTotal de Cost para la semana 12: {total_cost_semana_12:.2f}")
```
